## 1.最小二乘法

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#定义数据
x = np.arange(10)
y = x*3 + np.random.randint(1,5,size=10)
```

```python
#绘图
plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

#最小二乘法算法
k = np.sum((x-x_mean)*(y-y_mean))/np.sum((x-x_mean)**2)
b = y_mean - k*x_mean
print(k,b)
```

```python
y1 = k*x + b
plt.scatter(x,y)
plt.plot(x,y1)
plt.grid()
plt.show()
```

```python
def leastsq(x,y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    k = np.sum((x-x_mean)*(y-y_mean))/np.sum((x-x_mean)**2)
    b = y_mean - k*x_mean
    return {'k':k,'b':b}
```

```python
leastsq(x,y)
```

## 2.梯度下降算法

```python
x = np.linspace(-1,6,100)
y = (x-2.5)**2-1

plt.plot(x,y)
plt.show()
```

```python
#梯度下降算法
def J(x0):
    return (x0-2.5)**2-1
```

```python
x0 = 0.0
last_x = -0.1
eta = 0.4   #学习率
x_history=[]
#限制最大次数
i = 0
while i<100:
    dJ = J(x0)-J(last_x)
    if abs(dJ) < 1e-6:
        break
    last_x = x0
    x0 = x0 - eta*dJ
    x_history.append(x0)
    i+=1
print(x_history)
```

```python
plt.plot(np.array(x_history),J(np.array(x_history)),color='r',marker='+')
plt.plot(x,y)
plt.show()
```

```python
#共轭梯度下降法
```

```python
def df(x):
    return 2*(x-2.5)
```

```python
x0 = 0.0
eta = 0.9  #学习率
x_history=[]
#限制最大次数
i = 0
while i<100:
    dJ1 = df(x0)
    if abs(dJ1) < 1e-6:
        break
    x0 = x0 - eta*dJ1
    x_history.append(x0)
    i+=1
print(x_history)
```

```python
plt.plot(np.array(x_history),J(np.array(x_history)),color='r',marker='+')
plt.plot(x,y)
plt.show()
```
